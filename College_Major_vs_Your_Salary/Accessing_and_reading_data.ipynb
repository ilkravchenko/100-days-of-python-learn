```python
# What college major has the highest mid-career salary? How much do graduates with this major earn? 
# (Mid-career is defined as having 10+ years of experience).

max_mid_career_salary = clean_df['Mid-Career Median Salary'].max()
college_major_id = clean_df['Mid-Career Median Salary'].idxmax()
college_major = clean_df['Undergraduate Major'][college_major_id]
print(max_mid_career_salary)
print(college_major)
```

```python
# Which college major has the lowest starting salary and how much do graduates earn after university?

min_starting_salary = clean_df['Starting Median Salary'].min()
college_major_id = clean_df['Starting Median Salary'].idxmin()
college_major = clean_df['Undergraduate Major'][college_major_id]
print(min_starting_salary)
print(college_major)
```

```python
# Which college major has the lowest mid-career salary and how much can people expect to earn with this degree? 

min_mid_career_salary = clean_df['Mid-Career Median Salary'].min()
college_major_id = clean_df['Mid-Career Median Salary'].idxmin()
college_major = clean_df['Undergraduate Major'][college_major_id]
print(min_mid_career_salary)
print(college_major)
```
