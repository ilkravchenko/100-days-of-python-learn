```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
clean_df = df.dropna()
```

```python
clean_df.head()
```

```python
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# Using the .sort_values() method, can you find the degrees with the highest potential? 
# Find the top 5 degrees with the highest values in the 90th percentile. 

high_potencial = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_potencial[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference
# between high and low earners after graduation.

high_spread = clean_df.sort_values('Spread', ascending=False)
high_spread[['Undergraduate Major', 'Spread']].head()
```
